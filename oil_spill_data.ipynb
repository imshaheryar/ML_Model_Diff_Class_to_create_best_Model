```python
x = df.drop(['target'], axis = 1)
y = df['target']
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.25, random_state= 55)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
#fit the scaler to train set, it will learn the paerameters
scaler.fit(x_train)
#transform the train and test sets
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns = x_train.columns)
x_test_scaled = pd.DataFrame(x_test_scaled, columns = x_test.columns)
```

```python
x_train_scaled.head(3)
```

```python
x_test_scaled.head(3)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
def report(y_test, ypred):
    print('Confusion Matrix\n',confusion_matrix(y_test, ypred))
    print('Classification Report\n',classification_report(y_test, ypred))
    print('Accuracy Score\n',accuracy_score(y_test, ypred))
def score(model):
    print('Train Score', model.score(x_train, y_train))
    print('Test Score', model.score(x_test, y_test))
```

## LinearRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression(max_iter = 500, solver='liblinear')
model_lr.fit(x_train,y_train)
```

```python
ypred_lr = model_lr.predict(x_test)
print(ypred_lr[:5])
```

```python
report(y_test, ypred_lr)
```

```python
score(model_lr)
```

## KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_kn = KNeighborsClassifier(n_neighbors=10)
model_kn.fit(x_train,y_train)
```

```python
ypred_kn = model_kn.predict(x_test)
print(ypred_kn[:5])
```

```python
report(y_test, ypred_kn)
```

```python
score(model_kn)
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier(criterion='entropy', max_depth=4 , min_samples_split=25, random_state=45)
model_dt.fit(x_train,y_train)
```

```python
ypred_dt = model_dt.predict(x_test)
print(ypred_dt[:5])
```

```python
report(y_test,ypred_dt)
```

```python
#Overfit model
score(model_dt)
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators=100, criterion='entropy', min_samples_split=25, random_state=45)
model_rf.fit(x_train,y_train)
```

```python
ypred_rf = model_rf.predict(x_test)
print(ypred_rf[:5])
```

```python
report(y_test, ypred_rf)
```

```python
score(model_rf)
```

## BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model_bg = BaggingClassifier(base_estimator=model_dt, n_estimators=100, max_samples = x_train.shape[0], 
                             max_features=x_train.shape[1], random_state=45)
model_bg.fit(x_train, y_train)
```

```python
ypred_bg = model_bg.predict(x_test)
print(ypred_bg[:5])
```

```python
report(y_test, ypred_bg)
```

```python
score(model_bg)
```

## AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model_ab = AdaBoostClassifier(base_estimator = model_dt, n_estimators = 90, random_state=45)
model_ab.fit(x_train,y_train)
```

```python
ypred_ab = model_ab.predict(x_test)
print(ypred_ab[:5])
```

```python
report(y_test, ypred_ab)
```

```python
score(model_ab)
```
